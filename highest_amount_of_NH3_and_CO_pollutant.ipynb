```python
# Importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading in the dataset
df = pd.read_csv('/content/Air_Quality.csv')
df.head()
```

```python
# checking all the columns
df.columns
```

```python
pollutant = ['NH3' ,'NO2']
```

```python
# Indexing the rows which have the pollutant  'NH3'
df_nh3 = df.loc[df['pollutant_id'] == 'NH3']

# Indexing the rows which have the pollutant  'NO2'
df_no2 = df.loc[df['pollutant_id'] == 'NO2']
```

```python
df_nh3.head() 
```

```python
# getting the 10 rows with highest 'NH3' pollutant_avg
nh3_max = df_nh3.nlargest(10 ,['pollutant_avg'])
nh3_max
```

```python
# PLotting the 10 cities with highest 'NH3' pollutant_avg
plt.figure(figsize=(17,6))
plt.xticks(color = 'red' ,size = 12)
plt.yticks(color = 'red' ,size = 13)

ax = sns.barplot(x = nh3_max['city'] ,y = nh3_max['pollutant_avg'] 
            ,palette = 'Set1' ,edgecolor='black')

plt.xlabel('CITY' ,color = 'PURPLE',size = 15)
plt.ylabel('AVG  NH3  POLLUTANT' ,color = 'PURPLE' ,size = 15)

plt.grid(b = True ,
         linewidth = 0.5,
         linestyle = '--',
         color = 'gray')

for bar in ax.patches:
       ax.annotate( s = str(bar.get_height()) , # the text to put
                    xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    ha='center',
                    va='center',
                    size=14,
                    xytext=(0, 10),
                    textcoords='offset points')
       
for i,bar in enumerate(ax.patches):
       ax.annotate( s = nh3_max['state'].iloc[i], # the text to put
                    xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()/2),
                    ha='center',
                    va='center',
                    size=12,
                    xytext=(0, 10),
                    textcoords='offset points')   
plt.title("Cities with the highest NH3 pollutant_avg" ,size=17 ,color = 'pink')
```

```python
# getting the 10 cities with highest 'NO2' pollutant_avg
no2_max = df_no2.nlargest(10 ,['pollutant_avg']).reset_index()
no2_max
```
