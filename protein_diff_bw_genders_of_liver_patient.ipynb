```python
# Importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading in the dataset
df = pd.read_csv('data/indian_liver_patient.csv')
df.head()
```

```python
# Plotting Total protein difference between gender using a boxplot
plt.figure(dpi = 100 ,figsize = (14,7))
sns.boxplot(y=df['Total_Protiens'] , x=df['Gender'] ,palette = 'viridis')
plt.ylabel('Total Protein' ,size=12 ,color='violet')
plt.xlabel('Gender' ,size=12 ,color='violet')
plt.xticks(size=10 ,color='red')
plt.yticks(size=10 ,color='red');
```

```python
# Plotting the distribution on scatterplot
plt.figure(dpi=150)
sns.scatterplot(y=df['Total_Protiens'] , x=df['Age'] ,hue=df['Gender'] ,palette='Set1',size=7)
plt.ylabel('Total Protein' ,size=12 ,color='violet')
plt.xlabel('Gender' ,size=12 ,color='violet')
plt.xticks(size=10 ,color='red')
plt.yticks(size=10 ,color='red');
```
